```python
#employee salary
month=int(input("Enter the month number "))
dayspresent=int(input("Enter the number of days you were present "))
dailywage=2000
if (month==1 or month==3 or month==5 or month==7 or month==8 or month==10 or month==12):
  for i in range(0,1):
    if dayspresent>31:
      print("Enter valid number of days ")
    else:
      salary=(dailywage*dayspresent)
elif (month==4 or month==6 or month==9 or month==11):
  for i in range(0,1):
    if dayspresent>30:
      print("Enter valid number of days ")
    else:
      salary=(dailywage*dayspresent)
elif month==2:
  for i in range(0,1):
    if dayspresent>29:
      print("Enter valid number of days ")
    else:
      salary=(dailywage*dayspresent)
print(f"Your salary is {salary}")

#transaction
balance=salary
noofposstrans=100
for i in range(noofposstrans):
  valandops=input("Enter value and operation ") #200 D
  temp=[]
  temp=valandops.split()
  value=temp[0]
  operation=temp[1]
  val=int(value)
  if operation=='D' or operation=='d':
    balance=balance+val
  elif operation=='W' or operation=='w':
    if balance>0:
      balance=balance-val
    elif balance<=0:
      print("Insufficient funds available")
      break
  next=input("Do you want to continue with next transaction ")
  if next=='Y' or next=='y':
    continue
  elif next=='N' or next=='n':
    break
print(f"Your current balance is {balance}")
```
